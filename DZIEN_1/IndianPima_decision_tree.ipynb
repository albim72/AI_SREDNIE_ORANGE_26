```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
col_names = ['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age','label']
```

```python
pima = pd.read_csv('diabetes.csv',header=None,names=col_names)
pima.head()
```

```python
cols = ['glucose','bp','skin','insulin','bmi']
pima[cols] = pima[cols].replace(0,np.nan)
pima.fillna(pima.median(),inplace=True)
```

```python
feature_cols = ['pregnant','insulin','bmi','age','glucose','bp','pedigree']
```

```python
X = pima[feature_cols]
y = pima.label
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=12,splitter='best',criterion='gini')
clf = clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(f"Accuracy: {metrics.accuracy_score(y_test,y_pred)}")
```

```python
clfrnd = RandomForestClassifier(n_estimators=200,n_jobs=-1)
clfrnd.fit(X_train,y_train)
y_pred = clfrnd.predict(X_test)
print(f"Accuracy: {metrics.accuracy_score(y_test,y_pred)}")
```
